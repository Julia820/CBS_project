```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
ver = 'org'
```

```python
df = pd.read_csv(ver + '/cluster_' + ver + '_quality.csv')
df
```

```python
col = pd.unique(df['cluster'])
d={}
for i in col:
    d[i] = list(df[df['cluster']==i]['p_val_adj'])
m = max(len(d[i]) for i in d) 
d_2 = {}
for i in d:
    d_2[i] = d[i]+[0]* (m - len(d[i]))

genes = pd.DataFrame.from_dict(d_2)
genes
```

```python
matplotlib.rcParams["figure.figsize"] = [10, 14]
genes.hist(grid = False)
plt.savefig('clusters_qc_' + ver + '.png')
```

```python
df = pd.read_csv(ver + '/seurat_' + ver + '_cluster_ad.csv')
df.index = df.singler
```

```python
fig, axs = plt.subplots(5,6, figsize=(25, 25), facecolor='w', edgecolor='k')
axs = axs.ravel()
for i in df.cluster.unique():
    df.loc[df.cluster == i].plot.pie(y='count', figsize=(5, 5), labels = None, ax = axs[i], legend = False, ylabel = '')
    axs[i].set_title('Cluster '+ str(i))
axs[26].axis('off')
axs[27].axis('off')
axs[28].axis('off')
axs[29].axis('off')
plt.tight_layout()
plt.savefig(ver+'/clusters_pie.png')
```
